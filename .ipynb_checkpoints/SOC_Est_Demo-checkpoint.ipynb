```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate, linalg
import pickle
from SOC_Est_SPKF import SPKF_Data, SOC_Est_SPKF
```

```python
# -----------------------------------------------------------------------------------------------------------------
# The discharging part (script 1) of the dynamic test of P14 cell at 25 degC is used for demonstration.
# -----------------------------------------------------------------------------------------------------------------

# Specify cell and test parameters
cellID = 'P14'
T = 25
mag = 50
delta_t = 1 # sampling time interval

# Load raw testing data and proceed delta_t interpolation
dir = cellID + '_DYN'

if T > 0:
    file = '%s_%02d_P%02d_S1.xlsx' % (dir,mag,T)
else:
    file = '%s_%02d_N%02d_S1.xlsx' % (dir,mag,abs(T))
data = pd.read_excel(dir+'/'+file,'Channel_1-003')
data = data[['Test_Time(s)','Step_Index','Current(A)','Voltage(V)']]
data.columns = ['time','step','current','voltage']

# Keep step 2, and last 300s of step 1
t1 = data['time'][data['step'].index[data['step'] == 2][0]] - 300
t2 = data['time'].iloc[-1]
t = np.arange(t1, t2, delta_t)
i = -1*interpolate.interp1d(data['time'],data['current'])(t)
v = interpolate.interp1d(data['time'],data['voltage'])(t)
t = t - t[0]

# Calculate true soc
model_file = cellID + '_model.pickle'
Model = pickle.load(open(model_file,'rb'))

# -----------------------------------------------------------------------
# Calculate true SOC
# -----------------------------------------------------------------------
eta = Model.getParam('etaParam',T)

etai = np.zeros((len(v)))
for k in range(len(etai)):
    etai[k] = i[k]
    if i[k] < 0: etai[k] = etai[k]*eta

z = np.ones(len(etai)) - np.cumsum(np.concatenate(([0],etai[:-1])))/(Q0*3600)
```

```python
# -----------------------------------------------------------------------------------------------------------------
# Initiate BMS
# -----------------------------------------------------------------------------------------------------------------

# Load BMS paramters
Qbump = 5  # bumping factor

maxI = 30*Q0  # Detection range of current sensor (+/- 30C)
precisionI = 2**10  # Detetion precision of current sensor (10-bit)
q = 2*maxI/precisionI  # Resolution of current sensor
VarI = (q*delta_t)**2/12  # Current sensor noise is assumed to be the uniformly-distrubuted quantization error only

# Reserve storage for computed results
z_hat = np.zeros((len(v)))
Varz = np.zeros((len(v)))

# -----------------------------------------------------------------------
# Initaition for SOC_Est_SPKF
# -----------------------------------------------------------------------
# Initiate prior input
I0 = 0
Isign0 = 0

# Initiate prior state estimate 
iR0 = np.zeros(Model.numpoles)
h0 = np.zeros(1)
z0 = Model.SOCfromOCVtemp(v[0],T)

# Initiate prior state covariance 
VariR0 = 1e2
Varh0 = 1e-2
Varz0 = 1e-3
Covx0 = np.diag([VariR0,Varh0,Varz0])  

# Initiate process noise (Covw) and sensor noise (Covv)
Covw = np.diag([VarI])  # Uncertainty of current sensor, state equation
Covv = np.diag([5e-2])  # Uncertainty of voltage sensor, output equation
```

```python
# -----------------------------------------------------------------------------------------------------------------
# Proceed SOC estimate
# -----------------------------------------------------------------------------------------------------------------

# Set up SPKFdata
SPKFData = SPKF_Data()
SPKFData.setup(I0,Isign0,iR0,h0,z0,Covx0,Covw,Covv,Qbump)

# Iterate SOC estimatation
for k in range(len(v)):
    I = i[k];
    V = v[k];
    
    z_hat[k],Varz[k],SPKFData = SOC_Est_SPKF(T,I,V,delta_t,SPKFData,Model)
```

```python
plt.figure(figsize=(20, 8))
    
# plot SOC etimates
plt.subplot(1, 2, 1)
plt.plot(t/60,100*z,'b-',label='true')
plt.plot(t/60,100*z_hat,'g-',label='est')
plt.plot(t/60,100*(z_hat+3*Varz**0.5),'y--',label='bounds')
plt.plot(t/60,100*(z_hat-3*Varz**0.5),'y--')
plt.xlim(100,300)
plt.ylim(40,80)
plt.xlabel('Time (min)')
plt.ylabel('SOC (%)')
plt.title('Measured and Estimated SOC at %d$^\circ$C' % T)
plt.legend(loc='best')
    
# plot SOC estimation error
plt.subplot(1, 2, 2)
plt.plot(t/60,100*(z_hat-z),'r-', label='errors')
plt.plot(t/60,100*(3*Varz**0.5),'y--',label='bounds')
plt.plot(t/60,100*(-3*Varz**0.5),'y--')

plt.xlabel('Time (min)')
plt.ylabel('SOC Error(%)')
plt.title('SOC Estimation Errors at %d$^\circ$C' % T)
plt.legend(loc='best')
plt.show()
```
