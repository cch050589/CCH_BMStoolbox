```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from Q_Est_AWTLS import AWTLS_Data, Q_Est_AWTLS
```

```python
# -----------------------------------------------------------------------------------------------------------------
# Synthetic data is used for demonstration.
# -----------------------------------------------------------------------------------------------------------------

Q0 = 15  # initial cell capacity
slope = -0.001 # Q aging rate
m = 300  # update interval
n = 1000  # number of updates
gamma = 0.9  # fading memory factor

maxx = 0.8  # maximum SOC change
maxI = 30*Q0  # detection range of current sensor (+/- 30C)
precisionI = 2**10  # detetion precision of current sensor (10-bit)
q = 2*maxI/precisionI  # resolution of current snesor

Varx0 = (0.01)**2 + (0.01)**2  # variance of SOC change: Var(x) = Var(zk2-zk1) = Var(zk2) + Var(zk1)
Vary0 = (q/3600)**2*m/12  # varaince of integral current is assumed to be the accumulated quantization errors only

Varx = Varx0*np.ones(n)
Vary = Vary0*np.ones(n)

# --------------------------------------------------------------------
# Generate syntheic data
# --------------------------------------------------------------------
Q = Q0 + slope*np.arange(n)  # true cell capacity
x = np.random.rand(n)*2*maxx-maxx  # SOC change 
y = Q*x  # integral current
x = x + Varx**0.5*np.random.randn(n)  # noisy SOC change 
y = y + Vary**0.5*np.random.randn(n)  # noisy integral current
```

```python
# -----------------------------------------------------------------------------------------------------------------
# Proceed Q estimate
# -----------------------------------------------------------------------------------------------------------------

# Reserve storage for computed results
Q_hat = np.zeros(n)
VarQ = np.zeros(n)

# Set up AWTLSdata
AWTLSData = AWTLS_Data()
AWTLSData.setup(Q0,Varx0,Vary0,gamma)

# Iterate Q estimation
for k in range(n):
    Q_hat[k],VarQ[k],AWTLSData = Q_Est_AWTLS(x[k],y[k],Varx[k],Vary[k],AWTLSData)
```

```python
# plot Q estimates
plt.figure(figsize=(20, 5))
plt.plot(Q,'b-', label='true')
plt.plot(Q_hat,'g-', label='est')
plt.plot(Q_hat+3*VarQ**0.5,'y--',label='bounds')
plt.plot(Q_hat-3*VarQ**0.5,'y--')
plt.xlabel('update index')
plt.ylabel('Capacity (Ah)')
plt.title('Estimated Q')
plt.legend(loc='best')
plt.show()
```

```python
# Calcualte the RMS capcity estimate error
rmsQerr = np.mean((Q_hat-Q)**2)**0.5
print('RMS error =',rmsQerr,'Ah')
```
