```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
import pickle
from ESC_modeling import ESCmodel
```

```python
# -----------------------------------------------------------------------------------------------------------------
# The discharging part (script 1) of the dynamic test of P14 cell at 25 degC is used for demonstration.
# -----------------------------------------------------------------------------------------------------------------

# Specify cell and test parameters
cellID = 'P14'
T = 25
mag = 50
delta_t = 1 # sampling time interval

# Load raw testing data and proceed delta_t interpolation
dir = cellID + '_DYN'

if T > 0:
    file = '%s_%02d_P%02d_S1.xlsx' % (dir,mag,T)
else:
    file = '%s_%02d_N%02d_S1.xlsx' % (dir,mag,abs(T))
data = pd.read_excel(dir+'/'+file,'Channel_1-003')
data = data[['Test_Time(s)','Step_Index','Current(A)','Voltage(V)']]
data.columns = ['time','step','current','voltage']

# Keep step 2, and last 300s of step 1
t1 = data['time'][data['step'].index[data['step'] == 2][0]] - 300
t2 = data['time'].iloc[-1]
t = np.arange(t1, t2, delta_t)
i = -1*interpolate.interp1d(data['time'],data['current'])(t)
v = interpolate.interp1d(data['time'],data['voltage'])(t)
t = t - t[0]

# Load ESC model
model_file = cellID + '_model.pickle'
Model = pickle.load(open(model_file,'rb'))
```

```python
# -----------------------------------------------------------------------------------------------------------------
# Proceed cell simulation
# -----------------------------------------------------------------------------------------------------------------

# Initiate states
z0 = 1
iR0 = np.zeros((1,Model.numpoles))
h0 = 0

# Simulate cell voltage
vest,_,_ = Model.simCell(i,T,delta_t,z0,iR0,h0)
```

```python
plt.figure(figsize=(20,8))
    
# plot voltages
plt.subplot(1, 2, 1)
plt.plot(t/60,v,'b-', label='mea')
plt.plot(t/60,vest,'g--',label='est')
plt.xlabel('Time (min)')
plt.ylabel('Voltage (V)')
plt.title('Measured and Estimated Voltage at %d$^\circ$C' % T)
plt.legend(loc='best')
    
# plot modeling errors
plt.subplot(1, 2, 2)
plt.plot(t/60, v-vest,'r--',label='verr')
plt.xlabel('Time (min)')
plt.ylabel('Error (V)')
plt.title('Modeling Error at %d$^\circ$C' % T)
    
plt.show()
```

```python
# calaute the root mean squre error of SOC estimation
print('RMS simulation error =',np.mean((v-vest)**2)**0.5*1000,'mv')
```
